```python
import bs4 as bs #!pip install beautifulsoup4
import urllib.request
import re
```

```python
article = urllib.request.urlopen("https://es.wikipedia.org/wiki/Argentina").read()
```

```python
parsed_article = bs.BeautifulSoup(article, 'lxml') #!pip install lxml
```

```python
paragraphs = parsed_article.find_all('p') #Solo vamos a pedir los <p> párrafos de HTML
```

```python
article_text = ""

for p in paragraphs:
    article_text += p.text #Unimos todos los párrafos en un único string
```

```python
article_text = re.sub('\u200b|\u2082', ' ', article_text) #reemplazar \u200b
article_text = re.sub('\[\d+\]', ' ', article_text) #remover referencias del tipo [1]
article_text = re.sub('\[\w+\]', ' ', article_text) #remover referencias del tipo [e]
article_text = re.sub(r'\s+', ' ', article_text) #remover otros símbolos y espacios
```

```python
#Guardar en un txt
text_file = open("articulo.txt", "w", encoding="utf8")
n = text_file.write(article_text)
text_file.close()
```
